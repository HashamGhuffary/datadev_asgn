```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
import os
import pandas as pd
from PIL import Image
import numpy as np

# dataset
data_dir = 'sight_insight_dataset'

# Load CSV file containing image filenames and labels
train_df = pd.read_csv(os.path.join(data_dir, 'train.csv'))

# Initialize variables
img_height = 224
img_width = 224

# Function to load and preprocess image
def load_image(file_path):
    img = Image.open(file_path)
    img = img.resize((img_height, img_width))
    img_array = np.array(img) / 255.0  # Normalize pixel values
    return img_array

# Load and preprocess images and labels
X = [] 
y = train_df['diagnosis'].values

for filename, label in zip(train_df['id_code'], train_df['diagnosis']):
    subfolder = ['No_DR', 'Mild', 'Moderate', 'Severe', 'Proliferate_DR'][label]  # folder names of label
    file_path = os.path.join(data_dir, 'colored_images', subfolder, filename + '.png')
    img = load_image(file_path)
    X.append(img)

X = np.array(X) 
```

```python
print(X)
```

```python
print(y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# setup the Convolutional Neural Networks Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(5, activation='softmax')  
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Training
history = model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)
```

```python
model.save("identifier_model.h5")
```
